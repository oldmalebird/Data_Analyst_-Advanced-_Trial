# 五王之战分析 - 冰与火之歌

## 简介

五王之战（War of the Five Kings）是著名严肃奇幻小说《冰与火之歌》中的著名内战。这是一场规模空前、波及七大王国的内乱。顾名思义，前后共有五人在战争中称王：乔佛里、史坦尼斯、蓝礼均声称自己是铁王座的合法继承人。除此之外，罗柏·史塔克被北境众封臣推选为北境之王，巴隆·葛雷乔伊亦再度掀起独立大旗，欲摆脱铁王座的统治，自称为铁群岛之王。



本数据集(battles.csv)包含了五王之战期间的战争，它是所有战斗的大集合。该数据是Kaggle中[Game of Thrones](https://www.kaggle.com/mylesoneill/game-of-thrones)的一部分。

数据中的变量含义解释：
```
name: 战争的名称，字符变量。
year: 战争发生的年份，数值变量。
battle_number: 本数据中的unique id，对应每一场独立的战役，数值变量。
attacker_king: 攻击方的国王，"/"表示了国王的更换。例如："Joffrey/Tommen Baratheon"意味着Tomen Baratheon继承了Joffrey的王位，分类变量。
defender_king: 防守方的国王，分类变量。
attacker_1: 攻击方将领，字符变量。
attacker_2: 攻击方将领，字符变量。
attacker_3: 攻击方将领，字符变量。
attacker_4: 攻击方将领，字符变量。
defender_1: 防守方将领，字符变量。
defender_2: 防守方将领，字符变量。
defender_3: 防守方将领，字符变量。
defender_4: 防守方将领，字符变量。
attacker_outcome: 从攻击方角度来看的战争结果，分别有：win, loss, draw，分类变量。
battle_type: 战争的类别。pitched_battle: 双方军队在一个地点相遇并战斗，这也是最基本的战争类别；ambush: 以隐身或诡计为主要攻击手段的战争；siege: 阵地战；razing: 对未设防位置的攻击。分类变量。
major_death: 是否有重要人物的死亡，二进制变量。
major_capture: 是否有重要人物的被捕，二进制变量。
attacker_size: 攻击方力量的大小，并未对骑兵、步兵等士兵种类有所区分，数值变量。
defender_size: 防守方力量的大小，并未对骑兵、步兵等士兵种类有所区分，数值变量。
attacker_commander: 攻击方的主要指挥官。指挥官的名字中并没有包含头衔，不同的指挥官名字用逗号隔开，字符变量。
defender_commander: 防守方的主要指挥官。指挥官的名字中并没有包含头衔，不同的指挥官名字用逗号隔开，字符变量。
summer: 战争是否发生于夏天，二进制变量。
location: 战争发生的地点，字符变量。
region: 战争发生的地域，包括：Beyond the Wall, The North, The Iron Islands, The Riverlands, The Vale of Arryn, The Westerlands, The Crownlands, The Reach, The Stormlands, Dorne，分类变量。
note: 注释，字符变量。

```

## 数据评估和清理结果
该数据集有38个样本，样本数较少。原有25列数据，删除空值较多的5列（attacker_3, attacer_4, defender_3, defender_4, note ）后，剩余20列。

## 数据探索分析

**1.战争主要发生在何时、何地、何方？**

````python
#绘制每年发生战争次数的柱状图
year = df['year'].unique()
frequency = df.groupby('year')['year'].count().values
plt.figure(figsize = (6,8))
plt.bar(x = year, height = frequency, width = 0.5)
x_ticks = np.arange(297, 301, 1)
y_ticks = np.arange(0, 21, 5)
plt.xticks(x_ticks)
plt.yticks(y_ticks)
for a,b in zip(year, frequency):
    plt.text(a, b, b, ha='center', va= 'bottom')
plt.title('Fig 1. Battle Frequency')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show;
````

由Fig 1可知，五王之战跨时三年，战争共计38场，主要集中在第二年。

````python
#绘制每个区域发生战争次数的柱状图
region = df['region'].unique()
frequency = df.groupby('region')['region'].count().values
plt.figure(figsize = (12,8))
plt.bar(x = region, height = frequency, width = 0.5)
for a,b in zip(region, frequency):
    plt.text(a, b, b, ha='center', va= 'bottom')
plt.title('Fig 2. Battle Regions')
plt.xlabel('Region')
plt.ylabel('Frequency')
plt.show;
````

由Fig 2可知，五王之战主战场在The Crownlands和The North区域。

````python
#pivot table展示战争发生在各个国王之间的次数
df['count'] = 1
pd.pivot_table(df,index='attacker_king',columns = 'defender_king', values='count',aggfunc=np.sum)
````

由此Pivot Table可知，五王之战主要发生在Joffrey/Tommen Baratheon和Robb Stark之间，且两者互相攻击居多，其中：Joffrey/Tommen Baratheon进攻Robb Stark 10次，Robb Stark进攻Joffrey/Tommen Baratheon 9次。

**2.影响战争胜负的因素有那些？**

````python
#绘制战争发动方胜负次数的柱状图
attacker_outcome = df[df['attacker_outcome'].notnull()]['attacker_outcome'].unique()
frequency = df.groupby('attacker_outcome')['attacker_outcome'].count().values
plt.figure(figsize = (4,6))
plt.bar(x = attacker_outcome, height = frequency, width = 0.5)
for a,b in zip(attacker_outcome, frequency):
    plt.text(a, b, b, ha='center', va= 'bottom')
plt.title('Fig 3. Attacker Outcome')
plt.xlabel('Attacker Outcome')
plt.ylabel('Frequency')
plt.show;
````

````python
#了解战争类型
battle_type = df[df['battle_type'].notnull()]['battle_type'].unique()
print(battle_type)
frequency = df.groupby(['battle_type', 'attacker_outcome'])
frequency.size()
````

````python
#绘制战争类型与胜负关系的柱状图
battle_type = ['ambush', 'pitched battle', 'razing', 'siege']
#提取各类型战争输赢的次数，此处应该由我不知道的函数提取
win = [10, 10, 2, 10]
loss = [0, 4, 0, 1]
index = np.arange(len(win))
width = 0.4
plt.figure(figsize = (12, 8))
plt.bar(x = index, height = win, width = width, label = 'win',tick_label = battle_type,color = 'black')
plt.bar(x = index + width, height = loss, width = width, label = 'loss', color = 'grey')
plt.title('Fig 4. Battle Type vs Attacker Outcome')
plt.xlabel('Battle Type')
plt.ylabel('Frequency')
plt.legend()
plt.show();
````

由Fig 3和Fig 4可知，战争发起方胜率很大, 37场战争中赢了32场，胜率为86%；ambush战和razing战的胜率为100%（当然后者的样本数量只有2个，说服力不强），pitched battle的胜率相对最低，但14场战争中也有10场为发起战争方获胜。

````python
#删除attacker_size，defender_size和attacker_outcome列中有缺失数据的样本
battle_size_all = df[['attacker_size', 'defender_size', 'attacker_outcome']]
battle_size = battle_size_all.dropna(axis = 0, how = 'any')
battle_size.info()
````

````python
#绘制攻防双方兵力与战争胜负关系的散点图
colorList = []
for i in battle_size['attacker_outcome']:
    if i == 'win':
        colorList.append('green')
    else:
        colorList.append('red')
y_axis = np.arange(0, battle_size['defender_size'].max())
x_axis = np.arange(0, battle_size['defender_size'].max())
plt.figure(figsize = (12, 8))
plt.scatter(battle_size['attacker_size'], battle_size['defender_size'], color = colorList)
plt.title('Fig 5.Battle Size vs Attacker Outcome')
plt.xlabel('Attacker Size')
plt.ylabel('Defender Size')
plt.show()
````

在该数据集中，舍弃缺失双方兵力和胜负信息的样本后，仅剩15个样本。
Fig 5为作战双方兵力与输赢关系图，绿色代表攻方胜利，绿色代表攻方失败，从此图可看出，作战双方的兵力和战争胜负无显著相关性。

## 得出结论

五王之战共38场战争，发生于298至300年，其中299年集中爆发20场，主站区位于The Crownlands和The North区域。

参战国主要发生在Joffrey/Tommen Baratheon和Robb Stark之间，且两者互相攻击居多，这两个国王互相攻击的次数加起来为19次，占五王之战一半的战争数量。

发动战争方的胜率为86%；战争类型对胜负有一定影响：ambush战和razing战的胜率为100%；攻防双方的兵力与战争胜负无显著相关性。
